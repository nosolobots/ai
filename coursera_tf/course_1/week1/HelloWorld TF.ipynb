### Hello World of Deep Learning with NN

In this example, our NN is going to learn the relationship between two numbers. For example, something like the returned value of this function:

float hw_function(float x) {
    float y = (2 * x) - 1;
    return y;
}

In our case, the NN has to "learn" the relation between two series of numbers. The input values (x) and the expected returned values (y)

To accomplish that, whe need to feed the NN with a set of Xs and a set of Ys in order to the NN be able to figure out the relationship between them

---
#### Imports

First we need to import the required libraries: 

```python
import tensorflow as tf
import numpy as np
from tensorflow import keras
```

---
#### Define and Compile the NN

We are going to create the simplest NN. It will have 1 layer, and that layer will have 1 neuron, and the input shape to it is just 1 value (input=1, hidden=0, output=1)

```python
model = tf.keras.Sequential([keras.layers.Dense(units=1, input_shape=[1])])
```

Once defined our NN, we have to compile it. When we do so, we have to specify 2 functions, a loss and a optimizer.
<br><br>
The **loss** function measures the guessed answers against the known correct values. We'll use **mean squared error** in this case
<br><br>
The **optimizer** try to minimize the loss function based on how well it went. We'll use **stochastic gradient descent** in this case
<br><br>
The process will repeat for a number of **epochs**

```python
model.compile(optimizer='sgd', loss='mean_squared_error')
```

---
### Providing the Data

Next, we'll feed in some data. In this case, we're taking 6 Xs and 6 Ys
<br><br>
We'll use numpy to create the datasets

```python
xs = np.array([-1.0, 0.0, 1.0, 2.0, 3.0, 4.0])
ys = np.array([-3.0, -1.0, 1.0, 3.0, 5.0, 7.0])
```

---
### Training the NN

The next step is to train our NN using the previous data. From the provided data, it will make predictions of the expected results and it will try to learn to make better predictions, improving epoch by epoch (check loss)

```python
model.fit(xs, ys, epochs=500)
```

---
### Making predictions

Once we trained our NN to learn the relationship between X and Y, we can use it to figure out what would the the corresponding Y value for a input X

```python
print(model.predict([10.0]))
```

Pretty close!
<br><br>
Remember that NN deals with probabilities, not certainties. And, after all, we only provided a tiny bunch of numbers to learn!!
